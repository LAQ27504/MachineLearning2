```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
fig, ax= plt.subplots(figsize=(10,8))
sns.heatmap(data.corr(), cmap='Blues', annot=True, fmt='0.3f',ax=ax)
ax.set_title('Correlation Matrix of Dataset', fontsize=15);
```

```python
X_col= data.columns[:-1] 
Y_col= data.columns[-1]

X= data[X_col] 
Y= data[Y_col]
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
scaled_df = pd.DataFrame(scaled_data, columns=X.columns)

#PCA
pca = PCA()
pca_data = pca.fit_transform(scaled_data)

#Converting to dataframe
names = [f"PC{i+1}" for i in range(pca_data.shape[1])]
pca_df = pd.DataFrame(pca_data, columns=names)


print(pca_df.head())
print("-----------------------------------")
print("Shape of PCA DataFrame:", pca_df.shape)
```

```python
plt.figure(figsize=(10, 5))
for column in scaled_df.columns:
    plt.plot(scaled_df.index, scaled_df[column], alpha=0.6, label=column)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Adjusted Data')
plt.legend()
plt.grid()
plt.show()
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
PC_labels = [f"PC{i+1}" for i in range(len(explained_variance_ratio))]

#Plot the explained variance ratio
plt.bar(PC_labels, explained_variance_ratio, alpha=0.5, align='center')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio by Principal Component')
plt.show()
```

```python
eigenvalues = pca.explained_variance_

#Plot the scree plot of variance explained by eigenvalues
plt.figure(figsize=(10, 5))
plt.scatter(range(1, len(eigenvalues) + 1), eigenvalues, color='b')
plt.plot(range(1, len(eigenvalues) + 1), eigenvalues, color='b')
plt.xlabel('ith Eigenvector')
plt.ylabel('Variance explained by ith Eigenvector')
plt.title('Variance explained by the Eigenvectors')
plt.show()
```

```python
pca_highest = PCA(n_components=2)
pca_transformed = pca_highest.fit_transform(scaled_data)

#Convert to dataframe
pca_highest_df = pd.DataFrame(pca_transformed, columns=['PC1', 'PC2'])

explained_variance_sum = np.sum(pca_highest.explained_variance_ratio_)
print(f"\nTotal variance explained by PC1 and PC2: {explained_variance_sum:.4f} ({explained_variance_sum * 100:.2f}%)")

#Plot results
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_highest_df['PC1'], y=pca_highest_df['PC2'], alpha=0.7, edgecolor='black')
plt.axhline(0, color='blue', linewidth=2)
plt.axvline(0, color='orange', linewidth=2)
plt.title('PCA: 2D Projection using Two Highest Variance PCs')
plt.grid()
plt.show()
```

```python
pca = PCA()
pca_transformed = pca.fit_transform(scaled_data)

#Get sorted indices of PCs based on variance
sorted_indices = np.argsort(pca.explained_variance_)[::-1]

#Select and pick the highest and lowest variance principal components
highest_pc = sorted_indices[0]
lowest_pc = sorted_indices[-1]
selected_pc = [highest_pc, lowest_pc]
selected_data = pca_transformed[:, selected_pc]

explained_variance_sum_1 = np.sum(pca.explained_variance_ratio_[selected_pc])
print(f"\nTotal variance explained by PC{selected_pc[0]+1} and PC{selected_pc[1]+1}: {explained_variance_sum_1:.4f} ({explained_variance_sum_1 * 100:.2f}%)")

#Convert to dataframe
pca_selected_df = pd.DataFrame(selected_data, columns=['PC1', 'PC6'])

#Plot results
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_selected_df['PC1'], y=pca_selected_df['PC6'], alpha=0.7, edgecolor='black')
plt.axhline(0, color='blue', linewidth=2)
plt.axvline(0, color='orange', linewidth=2)
plt.title('PCA: 2D Projection using One Highest and One Lowest Variance PCs')
plt.grid()
plt.show()
```
