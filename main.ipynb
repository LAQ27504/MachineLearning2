```python
data_correlation = data.corr()
data_correlation
```

```python
import numpy as np

# Set correlation threshold
threshold = 0.8

# Find feature pairs with high correlation (ignoring self-correlation)
correlated_pairs = (
    data_correlation.where(np.triu(np.ones(data_correlation.shape), k=1) == 1)
    .stack()
    .reset_index()
)

# Filter only highly correlated pairs
correlated_pairs.columns = ['Feature 1', 'Feature 2', 'Correlation']
highly_correlated = correlated_pairs[abs(correlated_pairs['Correlation']) > threshold]

print(highly_correlated)
```
